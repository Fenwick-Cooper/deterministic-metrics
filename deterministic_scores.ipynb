```python
# Convert to flat arrays for percentile calculations and histograms
daily_precip_IMERG_flat = np.array(daily_precip_IMERG[np.logical_not(daily_precip_IMERG.mask)])
daily_precip_IMERG_flat = daily_precip_IMERG_flat[np.logical_not(np.isnan(daily_precip_IMERG_flat))]

daily_precip_IFS_flat = []
for lead_days in range(num_lead_days):
    daily_precip_IFS_flat.append(daily_precip_IFS[:,lead_days,:,:][np.logical_not(daily_precip_IFS[:,lead_days,:,:].mask)])
    daily_precip_IFS_flat[lead_days] = daily_precip_IFS_flat[lead_days][np.logical_not(np.isnan(daily_precip_IFS_flat[lead_days]))]
daily_precip_IFS_flat = np.array(daily_precip_IFS_flat)

daily_precip_cGAN_flat = []
for lead_days in range(num_lead_days):
    daily_precip_cGAN_flat.append(daily_precip_cGAN[:,lead_days,:,:][np.logical_not(daily_precip_cGAN[:,lead_days,:,:].mask)])
    daily_precip_cGAN_flat[lead_days] = daily_precip_cGAN_flat[lead_days][np.logical_not(np.isnan(daily_precip_cGAN_flat[lead_days]))]
daily_precip_cGAN_flat = np.array(daily_precip_cGAN_flat)
```

```python
# Histogram of rainfall
bins = np.arange(0,70.5,0.5)
plt.hist(daily_precip_IMERG_flat, bins, density=True,
         log=True, histtype='step', ec='r', lw=1, label="IMERG")
for lead_days in range(num_lead_days):
    plt.hist(daily_precip_IFS_flat[lead_days], bins,
             density=True, log=True, histtype='step', lw=1,
             label=f"IFS-30h")
             # label=f"IFS-{lead_days*24+6}h")
plt.xlabel("Rainfall (mm/day)")
plt.ylabel("Probability density")
plt.title(datetime(year,month,1).strftime("%B %Y"))
plt.legend()
plt.show()
```

```python
percentiles = np.arange(1,100)
plt.plot(np.percentile(daily_precip_IMERG_flat, percentiles), percentiles,'r--', label="IMERG")
for lead_days in range(num_lead_days):
    plt.plot(np.percentile(daily_precip_IFS_flat[lead_days,:], percentiles), percentiles, '-',
             # label=f"IFS-{lead_days*24+6}h")
             label=f"IFS-30h")
plt.ylabel('Percentiles')
plt.xlabel('Rainfall (mm/day)')
plt.title(datetime(year,month,1).strftime("%B %Y"))
plt.legend()
plt.show()
```

```python
min_rainfall = 0.1  # mm/day

precentiles = np.arange(1,100)
for lead_days in range(num_lead_days):
    x = np.percentile(daily_precip_IMERG_flat, precentiles)
    y = np.percentile(daily_precip_IFS_flat[lead_days], precentiles)
    plt.loglog(x, y, label=f"IFS-{lead_days*24+6}h")

# Carefully check the units?
plt.plot([0,max(x[-1],y[-1])], [0,max(x[-1],y[-1])], 'r--')

plt.xlim(min_rainfall, max(x[-1],y[-1])*1.1)
plt.ylim(min_rainfall, max(x[-1],y[-1])*1.1)
plt.xlabel('IMERG percentiles (mm/day)')
plt.ylabel('IFS percentiles (mm/day)')
plt.legend()
plt.title(datetime(year,month,1).strftime("%B %Y"))
plt.show()
```

```python
# Plot a single day
day=25

# Set the contour plot style
plot_levels, plot_colours = du.get_contour_levels('KMD')
plot_norm = 24  # We are plotting mm/day and plot_norm == 1 corresponds to mm/h

# Load the border shapefile
reader = shpreader.Reader(shapefile_name)
borders_feature = ShapelyFeature(reader.geometries(), ccrs.PlateCarree(), facecolor='none')

# Define the figure and each axes for the rows and columns
fig, axs = plt.subplots(nrows=1, ncols=3, subplot_kw={'projection': ccrs.PlateCarree()}, figsize=(16,6.5))

# axs is a 2 dimensional array of `GeoAxes`. Flatten it into a 1-D array
axs=axs.flatten()

ax=axs[0]  # First plot (left)
ax.add_feature(borders_feature, linewidth=0.5, edgecolor='black')  # The borders
ax.contourf(longitude,latitude,daily_precip_IMERG[day-1,:,:],
             colors=plot_colours, levels=plot_levels*plot_norm,
             transform=ccrs.PlateCarree())
ax.set_title('IMERG')
ax.set_frame_on(False)

ax=axs[1]  # Second plot (middle)
ax.add_feature(borders_feature, linewidth=0.5, edgecolor='black')  # The borders
ax.contourf(longitude,latitude,daily_precip_IFS[day-1,0,:,:],
             colors=plot_colours, levels=plot_levels*plot_norm,
             transform=ccrs.PlateCarree())
ax.set_title('IFS')
ax.set_frame_on(False)

ax=axs[2]  # Third plot (right)
ax.add_feature(borders_feature, linewidth=0.5, edgecolor='black')  # The borders
c = ax.contourf(longitude,latitude,daily_precip_cGAN[day-1,0,:,:],
                colors=plot_colours, levels=plot_levels*plot_norm,
                transform=ccrs.PlateCarree())
ax.set_title('cGAN')
ax.set_frame_on(False)

# Adjust the location of the subplots on the page to make room for the colorbar
fig.subplots_adjust(bottom=0.1, top=0.9, left=0.1, right=0.8,
                    wspace=0.02, hspace=0.02)

# Add a colorbar axis at the bottom of the graph
cbar_ax = fig.add_axes([0.81, 0.215, 0.015, 0.57])

# Draw the colorbar
cb=fig.colorbar(c, cax=cbar_ax,orientation='vertical')
#cb.ax.tick_params(labelsize=10)
cb_labels = np.round(plot_levels*plot_norm,1).astype(str).tolist()
cb_labels[-1] = ''  # Remove the final label
cb.set_ticks(ticks=plot_levels*plot_norm, labels=cb_labels)
cb.set_label('Rainfall (mm/day)')

title_string = f"""Ensemble mean, lead time: 30h
Valid {datetime(year,month,day) + timedelta(hours=6)} to {datetime(year,month,day) + timedelta(hours=6+24)}"""
fig.suptitle(title_string)
plt.show()

# XXX Add space for WRF-KMD, WRF-EMI, IMERG climatology, ERA5, ICPAC-WRF climatology, Fuxi, GraphCast.
# XXX Move colorbar to the bottom
```
